# Rad u Jyputer-u

Džupiter (Jupyter) je interaktivna radna sveska u koju možete da unosite tekst (kao tekst koji upravo čitate), da računate, da pokrećete jednostavnije Pajton programe, da obrađujete podatke, da prikazujete podatke u vidu tabele i dijagrama, i još mnogo toga.
Džupiter interaktivne radne sveske se mogu pokrenuti na više načina, a mi ćemo ga za izučavanje koristiti pomoću servisa Binder. Ovaj servis je dostupan na adresi mybinder.org (ne treba ništa instalirati, ne treba kreirati nikakav nalog).
Najjednostavniji način da otvorite Džupiter interaktivne radne sveske koje prate naše časove informatike i računarstva je da odete na link koji će vam biti prosleđen.


##### Ovo je izgled sajta koji ćete pokrenuti:

Servis će napraviti novu kopiju svih radnih sveski koje prate ovaj kurs i pripremiće okruženje u kome može da se izvršava Džupiter. Priprema traje (treba malo strpljenja).Prepoznaćete da je sve gotovo kada se u veb pregledaču (*web browser*) na vašem računaru otvori stranica na kojoj se vide linkovi na radne sveske. Кlikom na ime, radna sveska se otvara u novom tabu veb pregledača (kao na slici ispod)
