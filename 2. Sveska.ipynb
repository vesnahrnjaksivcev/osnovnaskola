## **Upotreba promenljivih i Pajton komandi u Džupiteru** 

**1. Zadatak** 

Milica je želela da kupi pantalone koje su koštale 6.799,99 dinara, ali joj je mama rekla da su preskupe. Zato je Milica sačekala sezonu sniženja i kada je sledeći put obišla prodavnicu videla je da je cena pantalona snižena za 25%. Mama je pristala da joj kupi pantalone po sniženoj ceni. Кada su došle do kase prijatno su bile iznenađene činjenicom da su dobile popust od 3% na već sniženu cenu zato što je njena mama pantalone platila gotovinom. Кoliko su na kraju koštale pantalone?

**2. Zadatak**

(a) Jedna knjiga je koštala 2460 dinara, a onda je knjižara snizila njenu cenu za 20%. Кoliko sada košta knjiga?

(b) Druga knjiga u istoj knjižari nakon sniženja od 20% košta 1480 dinara. Кoliko je koštala pre sniženja?

**3. Zadatak**

Cena litra jogurta od 72 din povećana je za 12%. Kolika je nova cena jogurta?

**4. Zadatak**
Prodavnica nameštaja daje letnji popust u iznosu od 15%, pa je cena francuskog ležaja sa popustom 23800 dinara.
Kolika je bila cena pre pojeftinjenja i koliko iznosi pojeftinjenje?

**5. Zadatak** 

Procenjuje se da je Кina na dan 1.7.2019. imala 1420062022 stanovnika. Broj stanovnika u Кini se svake godine poveća za 0,35%. Pod pretpostavkom da se godišnji rast broja stanovnika u Кini neće menjati, koliko stanovnika će Кina imati 2020. godine?
