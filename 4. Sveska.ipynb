**<font color= red >PONAVLJANJE (FOR PETLJA)</font>**

**Ponavljanje naredbi dati broj puta:**
Razmislimo kako bismo napisali program koji deset puta ispisuje Zdravo. Na osnovu onoga što smo do sada naučili možemo deset puta navesti naredbu print("Zdravo").

**<font color= red >PETLJE</font>** su naredbe koje omogućavaju da se neka naredba ili neki blok naredbi ponove više puta. Za početak razmotrimo naredbu kojom postižemo da se nešto ponovi dati broj puta.


```python
for i in range(10):                 # ponovi 10 puta:
    print("Zdravo")                 # ispiši tekst "Zdravo"
```

Dаkle, komаndа kojа se koristi zа ponovljeno izvršаvаnje komаnde print jeste nаredbа **<font color= red >FOR</font>**.
Ukoliko bi želeli da se naredba print, u ovom slučaju „Zdravo“ ispiše u jednom redu, a ne u odvojenim redovima, uz print bi dodali i **<font color= red >end=(" ") </font>** pa bi zadatak izgledao ovako: 

Komanda gde se **i** nаzivа brojаčem koji uzimа vrednosti iz intervаlа od 0 do, u ovom slučаju 9, koji je definisаn komаndom **range(10)**. Dа bi se Komanda ponovilа 50 putа potrebno je u funkciji **range** umesto 10 nаvesti 50.

**<font color= red >Brojačke promenljive:</font>**
Tokom izvršavanja petlje oblika for i in range(n): promenljiva i redom uzima vrednosti 0, 1, 2 itd., sve do vrednosti n-1. Na primer, ako je n jednako 3, tada će se telo petlje izvršiti četiri puta i u prvom izvršavanju tela petlje promenljiva i će imati vrednost 0, u drugom vrednost 1, a u trećem vrednost 2.
Ispis brojeva od 0 do 9

**<font color= red >Ispis brojeva od a do b:</font>**
Kada se u pozivu funkcije range navede samo jedan argument, onda brojanje kreće od nule, međutim, kada se navedu dva argumenta a i b, tada brojanje kreće od a i završava se sa b-1.
Napisati program koji učitava dva broja a i b i zatim ispisuje sve brojeve od a do b (uključujući i njih).

Ako se u pozivu funkcije range navedu tri argumenta, treći od njih određuje korak za koji se u svakom koraku petlje vrši uvećavanje brojačke promenljive.
Na primer, range(1, 10, 2) označava da brojačka promenljiva kreće od 1 i da se u svakom koraku uvećava za 2, sve dok je strogo manja od 10 tj. da redom uzima vrednosti 1, 3, 5, 7 i 9.

Direktor fabrike ima spisak plata radnika:

Potreban mu je:

•	Spisak plata radnika sortiran od najveće do najmanje plate,

•	Prosečna plata svih radnika u fabrici

•	Sortiran spisak plata uvećanih za 5%.


```python
plate=[32000,34000,33000,29500,28700,35000,59510,45000,37000]
```

Kako spisak plata treba da bude poređan od najveće do najmanje , za sortiranje lemenata niza upotrebiće se **plate.sort(reverse=True)**

Prosečna plata radnika će se izračunati tako što se zbir svih elemenata podeli sa dužinom, niza. To se može uraditi uvođenjem promenljive **prosek=sum(plate)/len(plate)**. I prosečna plata radnika se može ispisati odgovarajućim teksto koristeći funkciju **print**.

Obzirom da je potreban sortiran spisak plata uvećanih za 5%, jedan od načina na koji se elementi niza mogu pojedinačno uvećati za 5% jeste pomoću **for petlje**. Na prvom mestu parametara opsega for petlje staviće se vrednost 0 i kao drugi dužina niza len(plate) – što znači da će biti uključeni svi elementi niza, dok se kroz print unutar for petlje zadaje naredba da se svim elementima plata uveća za 5% .

Ukoliko bi želeli da sačuvamo niz plata koje su uvećane za 5%, morali bi da kreiramo novi niz koji će sadržati plate radnika uvećane za 5% kao na slici ispod.
