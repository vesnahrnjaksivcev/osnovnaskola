**<font color= red > MEDIJANA NIZA</font>**

Aritmetička sredina je dobar pokazatelj kada su podaci ravnomerno rapoređeni.
Međutim, ako su podaci sa kojima radimo „veoma razbacani“ (imaju velike skokovite vrednosti odstupanja) može se desiti da prosek nije dobar predstavnik celog niza.
Na primer: uzmimo u obzir prosečne prihode građana jednog grada. Pošto se sve vrednosti zbrajaju, a zatim dele sa sa brojem građana, prihod građanina sa izuzetno visokim prihodom bi značajno uticao na srednju vrednost. Tako da možemo da zaključimo da aritmetička sredina nije uvek dobar prikaz podataka.  Zato se koristi medijana niza. Iz sortiranog niza u rastućem poretku, se uzima element  u sredini.


1.	Ako je dužina niza neparna medijana niza je element tačno na sredini niza.
2.	Ako je dužina niza parna uzimamo aritmetičku sredinu dva niza srednja broja u nizu.

Za vizuelni prikaz podataka se koristi biblioteka **<font color= red >matplotlib.pyplot</font>**. Biblioteka se uključuje u program naredbom **<font color= red >import</font>** i obično joj se dodeljuje skraćeni naziv **<font color= red >plt</font>**, kako bi zapisivanje naredbi bilo jednostavnije.


Za kreiranje grafikona su potrebna najmanje dva niza, od kojih će jedan biti na x-osi, a drugi na y-osi.

**<font color= red > KREIRANJE LINIJSKOG GRAFIKONA </font>**
