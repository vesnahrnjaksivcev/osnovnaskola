**<font color= red > Grafičko predstavljanje – histogram </font>**

Za vizuelni prikaz podataka se koristi biblioteka **<font color= red >matplotlib.pyplot</font>**. Biblioteka se uključuje u program naredbom **<font color= red >import</font>** i obično joj se dodeljuje skraćeni naziv **<font color= red >plt</font>**, kako bi zapisivanje naredbi bilo jednostavnije.

Za kreiranje grafikona su potrebna najmanje dva niza, od kojih će jedan biti na x-osi, a drugi na y-osi.


**<font color= red > Kreiranje stubičastog grafikona – histograma (engl. bar charts)</font>**

Ovaj dijagram je veoma pogodan za prikazivanje niza u kom nema mnogo podataka.
Za kreiranje stubičastog grafikona se koristi funkcija **<font color= red >bar</font>** iz biblioteka **<font color= red >matplotlib.pyplot</font>**.

Da bi se promenila veličina grafikona, koristi se funkcija **<font color= red >plt.figure(figsize=(širina,visina))</font>** . Funkcija za podešavanje veličine grafikona se uvek poziva pre kreiranja grafikona.
Da bi se stubić uokvirio nekom bojom koristi se parametar **<font color= red > edgecolor="boja"</font>** .
Promena širine stubića se radi pomoću parametra **<font color= red > width=širina </font>** , standardna širina je 0.8.
